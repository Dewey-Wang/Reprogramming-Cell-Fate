```r
library(readr, quietly = TRUE, warn.conflicts = FALSE)
library(Seurat, quietly = TRUE, warn.conflicts = FALSE)
```

```r
data = read_rds('../data/generated data from script/data.rds')
bootstrap_clustering = read_rds('../data/generated data from script/bootstrap_clustering.rds')
robustness_clustering = read_rds('../data/generated data from script/robustness_clustering.rds')
cluster_3_HVG_HLG_genes_list = read_rds('../data/generated data from script/cluster_3_HVG_HLG_genes_list.rds')
df = read_rds('../data/generated data from script/df.rds')

vedo=read_rds('../data/xenopus_combined_and_clustered_no_out.rds')
vedo = UpdateSeuratObject(object = vedo)
subset_data <- subset(vedo, idents = 9)
subset_data <- SCTransform(subset_data,  variable.features.n = 3000) # SCTransform：适合在小数据集或高噪声情况下使用，结果通常更可靠。 对于目标 cluster 较小的情况 (如占比 4%)，SCTransform 更适合，尤其是它能更好地处理技术噪声。

# 3. 选择 top 10 HVGs
cluster_9_hvgs <- head(VariableFeatures(subset_data), 10)

# 打印 top 10 HVGs
print(cluster_9_hvgs)
```

```r
data = t(data)
data = merge(x = data, y = robustness_clustering, by.x = "row.names", by.y = "Gene", all.x = TRUE)
rownames(data) <- data$Row.names
# 2. 移除 Row.names 欄位 (因為它現在是 row names，不再需要作為一個單獨的欄位)
data$Row.names <- NULL
data <- data[ , !(colnames(data) %in% "Cluster")]
# 顯示修改後的 data frame
print(data)
```

```r
length(colnames(data))
```

```r
divide_with_max_gene <- data.frame()
for (i in 1:length(row.names(data))) {
  max_level <- max(data[i,])
  for (a in 1:(length(data)-1)) {
      divide_with_max_gene[i,a] <- data[i,a] / max_level
    }
}
divide_with_max_gene[,135]<-data[,135]
colnames(divide_with_max_gene) = colnames(data)
row.names(divide_with_max_gene) = row.names(data)
divide_with_max_gene <- t(divide_with_max_gene)
```

```r
group_1 <- divide_with_max_gene[,divide_with_max_gene['MappedCluster',]==1]
group_1 <- group_1[1:134,]
group_1<-cbind.data.frame(group_1,as.data.frame(df$dptval))
colnames(group_1)[length(group_1)] <- 'dptval'

group_2 <- divide_with_max_gene[,divide_with_max_gene['MappedCluster',]==2]
group_2 <-group_2[1:134,]
group_2<-cbind.data.frame(group_2,as.data.frame(df$dptval))
colnames(group_2)[length(group_2)] <- 'dptval'
```

```r
group_1_summary <- data.frame()
for (i in 1:(nrow(group_1))) {
  group_1_summary[i,1]<- mean(as.numeric(group_1[i,1:length(group_1)-1]))
  group_1_summary[i,2]<- sd(as.numeric(group_1[i,1:length(group_1)-1]))  
}
colnames(group_1_summary) = c('mean', 'sd')

group_1_summary$dptval <- df$dptval
group_1_summary
```

```r

library(ggplot2)
g1 <- ggplot(group_1_summary) + 
  stat_smooth(aes(x = dptval, y = mean-sd), method = "loess", se = FALSE, fill = "grey",alpha = 0.4) +
  stat_smooth(aes(x = dptval, y = mean+sd), method = "loess", se = FALSE, fill = "grey",alpha = 0.4)

# build plot object for rendering 
gg1 <- ggplot_build(g1)

# extract data for the loess lines from the 'data' slot
df2 <- data.frame(x = gg1$data[[1]]$x,
                  ymin = gg1$data[[1]]$y,
                  ymax = gg1$data[[2]]$y) 

# use the loess data to add the 'ribbon' to plot 
p1<-ggplot(group_1_summary) +
  geom_ribbon(data = df2, aes(x = x, ymin = 0, ymax = ymax),
              fill = "blue", alpha = 0.05)+
    stat_smooth(aes(x = dptval, y = mean), method = "loess", se = FALSE)+
labs(x= "Diffusion pseudotime", y = "gene expression") + ggtitle(paste("Group 1 - selected"
           ,length(group_1)-1
           ,"genes")) +   theme_classic() +
  theme(plot.title = element_text(hjust = 0.5))
```

# plot selected genes

The genes are selected from notebook `2. identify interest cluster.ipynb`

```r
manual_interest_gene <- c(cluster_9_hvgs, cluster_3_HVG_HLG_genes_list$HVG[1:10], cluster_3_HVG_HLG_genes_list$HLG[1:10])

manual_interest_gene_norm <- data.frame((divide_with_max_gene[,colnames(divide_with_max_gene) %in% manual_interest_gene ]))
colnames(manual_interest_gene_norm) <- colnames(divide_with_max_gene)[colnames(divide_with_max_gene) %in% manual_interest_gene]

for (i in 1:length(row.names(manual_interest_gene_norm))) {
  for(a in 1:length(row.names(df))){
    if (row.names(manual_interest_gene_norm)[i] == row.names(df)[a]){
      manual_interest_gene_norm$dpt[i] = df$dptval[a]
    }
  }
}

manual_interest_gene_norm = manual_interest_gene_norm[order(manual_interest_gene_norm$dpt), ]

merge_data_into_long = function (base_data, merged_data) 
{
  out_data_name = data.frame(x = rep(base_data,times=length(merged_data)))
  b = 1
  for (i in 1:length(merged_data)) {
    for (a in 1:length(row.names(merged_data))){
      out_data_name[b,2] = merged_data[a,i]
      out_data_name[b,3] = colnames(merged_data[i])
      b = b + 1
    }
  }
  colnames(out_data_name) = c('x', 'y','id')
  return(out_data_name)
}


test = merge_data_into_long(manual_interest_gene_norm$dpt,manual_interest_gene_norm[1:21])
```

```r

# 将基因复制到多个类别
expand_categories <- function(gene_id) {
  categories <- c()
  if (gene_id %in% cluster_9_hvgs) categories <- c(categories, "Cluster 9 HVG")
  if (gene_id %in% cluster_3_HVG_HLG_genes_list$HVG) categories <- c(categories, "Cluster 3 HVG")
  if (gene_id %in% cluster_3_HVG_HLG_genes_list$HLG) categories <- c(categories, "Cluster 3 HLG")
  
  # 为每个类别生成一行数据
  data.frame(id = gene_id, Category = categories, stringsAsFactors = FALSE)
}

# 为每个基因生成所有对应类别的行
expanded_test <- do.call(rbind, lapply(unique(test$id), expand_categories))

# 合并回原始数据
test_expanded <- merge(test, expanded_test, by = "id")

# 标记重复基因
test_expanded$Repeated <- test_expanded$id %in% expanded_test$id[duplicated(expanded_test$id)]

library(dplyr)

library(ggplot2)
library(patchwork)
# 示例数据
df <- test_expanded

# 更新背景颜色数据集
hacky_df <- df %>%
  distinct(id, Category) %>%
  mutate(
    facet_color = case_when(
      id %in% df$id[df$Repeated] ~ "repeated",  # 重复基因背景为红色
      TRUE ~ "normal"  # 非重复基因背景为白色
    )
  )

# 创建图表列表
plots <- list()
df$Category <- factor(df$Category, levels = c("Cluster 9 HVG", "Cluster 3 HVG", "Cluster 3 HLG"))

# 针对每个 Category 绘制单独的图表
for (cat in levels(df$Category)) {
  # 筛选当前 Category 数据
  data_subset <- df[df$Category == cat, ]
  
  # 根据当前 Category 的基因顺序重新排序
  if (cat == "Cluster 9 HVG") {
    current_order <- cluster_9_hvgs
  } else if (cat == "Cluster 3 HVG") {
    current_order <- cluster_3_HVG_HLG_genes_list$HVG[1:10]
  } else if (cat == "Cluster 3 HLG") {
    current_order <- cluster_3_HVG_HLG_genes_list$HLG[1:10]
  } else {
    current_order <- unique(data_subset$id)
  }
  
  # 转换 `id` 为 factor 并设置级别顺序
  data_subset$id <- factor(data_subset$id, levels = current_order)
  
  # 筛选背景颜色数据
  hacky_subset <- hacky_df[hacky_df$Category == cat, ]
  hacky_subset$id <- factor(hacky_subset$id, levels = current_order)  # 确保 hacky_subset 和 data_subset 的因子级别一致
  
  # 绘制图表
  p <- ggplot(data_subset) +
    # 添加背景矩形，动态设置每个分面的背景颜色
    geom_rect(
      data = hacky_subset,
      aes(
        xmin = -Inf, xmax = Inf,
        ymin = -Inf, ymax = Inf,
        fill = facet_color
      ),
      inherit.aes = FALSE,
      alpha = 0.2  # 设置透明度
    ) +
    geom_point(aes(x = x, y = y), color = 'black', shape = 21, size = 1) +
    geom_smooth(aes(x = x, y = y), color = 'orange') +  # 平滑线改为橙色
    facet_wrap(~ id, scales = "free", nrow = 2) +  # 按基因分面，确保使用因子顺序
    scale_fill_manual(
      values = c(
        "repeated" = "red",  # 重复基因背景色为红色
        "normal" = "white"   # 非重复基因背景色为白色
      )
    ) +
    theme_bw() +
    theme(
        strip.background = element_rect(fill = NA),  # 去除分面默认背景色
        strip.text = element_text(face = "bold", size = 20),     # 分面标题加粗
        plot.title = element_text(size = 30, face = "bold"),  # 设置标题样式和大小
        # 修改x/y轴标签样式
        axis.title.x = element_text(size = 16, face = "bold"),  
        axis.title.y = element_text(size = 16, face = "bold"),
        # 修改图例文本和标题样式
        legend.text = element_text(size = 16),  
        legend.title = element_text(size = 16),  
    ) +
    labs(
      title = paste("Category:", cat),  # 每个类别显示对应标题
      x = "Diffusion pseudotime",
      y = "Gene expression"
    )
  
  # 保存图表到列表
  plots[[cat]] <- p
}

# 使用 patchwork 将所有类别的图表合并
library(patchwork)
final_plot <- wrap_plots(plots, ncol = 1)
options(repr.plot.width = 23, repr.plot.height = 15, repr.plot.res = 100)

# 显示图表
print(final_plot)
```
